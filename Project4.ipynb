```python
#Print all unique values
null['Education'].unique()
```

```python
 # Transform categorical values to numerical
def changeeducation(edu):
    if edu == "Graduation":
        return 1
    elif edu == 'PhD':
        return 2
    elif edu == 'Master':
        return 3
    elif edu == 'Basic':
        return 4
    else:
        return 0


null["Education"] = null["Education"].apply(changeeducation)
null.head()
```

```python
#Check numerical values
null['Education'].unique()
```

```python
#Print unique values
null['Marital_Status'].unique()
```

```python
 # Transform categorical values to numerical
def changestatus(marital):
    if marital == "Single":
        return 1
    elif marital == 'Together':
        return 2
    elif marital == 'Married':
        return 3
    elif marital == 'Divorced':
        return 4
    elif marital == 'Widow':
        return 5
    elif marital == 'Alone':
        return 6
    elif marital == 'Absurd':
        return 7
    else:
        return 0

null["Marital_Status"] = null["Marital_Status"].apply(changestatus)
null.head()
```
