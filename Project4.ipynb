```python
null['Education'].unique()
```

```python
 # Transform Channel
def changeeducation(edu):
    if edu == "Graduation":
        return 1
    elif edu == 'PhD':
        return 2
    elif edu == 'Master':
        return 3
    elif edu == 'Basic':
        return 4
    else:
        return 0

# Along with replace() and map(), this is another way to encode the gender column into numbers.
null["Education"] = null["Education"].apply(changeeducation)
null.head()
```
