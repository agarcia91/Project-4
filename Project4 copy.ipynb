```python
#Import dependencies
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
import sklearn as skl
import tensorflow as tf
from sklearn.model_selection import train_test_split
import keras_tuner as kt
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
#Read the CSV file
df = pd.read_csv('/Users/anthonygarcia/Desktop/project_4/marketing_campaign.csv')
df
```

```python
# Find null values
for column in df.columns:
    print(f"Column {column} has {df[column].isnull().sum()} null values")
 
```

```python
# Drop all null values
new_df = df.dropna()
new_df
```

```python
# Find duplicate entries
print(f"Duplicate entries: {new_df.duplicated().sum()}")
```

```python
#Print all unique values
new_df['Education'].unique()
```

```python
 # Transform categorical values to numerical
def changeeducation(edu):
    if edu == "Graduation":
        return 1
    elif edu == 'PhD':
        return 2
    elif edu == 'Master':
        return 3
    elif edu == 'Basic':
        return 4
    else:
        return 0


new_df["Education"] = new_df["Education"].apply(changeeducation)
new_df.head()
```

```python
#Check numerical values
new_df['Education'].unique()
```

```python
#Print unique values
new_df['Marital_Status'].unique()
```

```python
 # Transform categorical values to numerical
def changestatus(marital):
    if marital == "Single":
        return 1
    elif marital == 'Together':
        return 2
    elif marital == 'Married':
        return 3
    elif marital == 'Divorced':
        return 4
    elif marital == 'Widow':
        return 5
    elif marital == 'Alone':
        return 6
    elif marital == 'Absurd':
        return 7
    else:
        return 0

new_df["Marital_Status"] = new_df["Marital_Status"].apply(changestatus)
new_df.head()
```

```python
# Check new values
new_df['Marital_Status'].unique()
```

```python
# Drop columns
_df = new_df.drop(['Dt_Customer'], axis=1)
_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy_df = pd.get_dummies(_df)
dummy_df
```

```python
# Train the linear regression model
y = dummy_df["Response"]
X = dummy_df.drop(["Response"],axis=1)
rf = LinearRegression()
rf.fit(X, y)
y_pred = rf.predict(X)
```

```python
# print score
r2_score(y, y_pred)
```

```python
 # Split our preprocessed data into our features and target arrays
y = dummy_df["Response"]
X = dummy_df.drop(["Response"],axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
 # Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))


# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
